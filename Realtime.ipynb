```python
import cv2
from keras.models import model_from_json
import numpy as np
```

```python
json_file = open("emotiondetector1.json", "r")
model_json = json_file.read()
json_file.close()
model = model_from_json(model_json)
```

```python
model.load_weights("emotiondetector1.h5")
haar_file=cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
face_cascade=cv2.CascadeClassifier(haar_file)
```

```python
def extract_features(image):
    feature = np.array(image)
    feature = feature.reshape(1,48,48,1)
    return feature/255.0
```

```python
# Initialize webcam
webcam = cv2.VideoCapture(0)

# Emotion labels
labels = {0: 'angry', 1: 'disgust', 2: 'fear', 3: 'happy', 4: 'neutral', 5: 'sad', 6: 'surprise'}

while True:
   
    i, im = webcam.read()
    if not i:
        break 

    gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)  # Detect faces

    for (p, q, r, s) in faces:
        image = gray[q:q + s, p:p + r]  # Extract face region
        cv2.rectangle(im, (p, q), (p + r, q + s), (255, 0, 0), 2)

        image = cv2.resize(image, (48, 48))
        img = extract_features(image) 

        pred = model.predict(img)
        prediction_label = labels[pred.argmax()] 

        # Display emotion text
        cv2.putText(im, '%s' % prediction_label, (p - 10, q - 10),
                    cv2.FONT_HERSHEY_COMPLEX_SMALL, 2, (0, 0, 255))

    cv2.imshow("Output", im)  # Show webcam feed

    # Break loop when 'x' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break


webcam.release()
cv2.destroyAllWindows()
```
